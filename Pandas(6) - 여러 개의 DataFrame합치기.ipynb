```python
# DataFrame을 합치는 방법은 2가지가 있다.
# 1. concatenating(연결)
# 2. merging(병합)
```

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({'key': list('bbacaab'),
                                  'data1': range(7)})
df2 = pd.DataFrame({'key': list('abd'),
                                   'data2': range(3)})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, on='key')
```

```python
pd.merge(df1, df2, on="key", how="outer")
```

```python
pd.merge(df1, df2, on='key', how='left')
```

```python
pd.merge(df1, df2, on='key', how='right')
```

```python
df1 = pd.DataFrame({'key': list("bbacab"),
                                   'data1': range(6)})
df2 = pd.DataFrame({'key': list("ababd"),
                                   'data2': range(5)})
```

```python
df1
```

```python
df2
```

```python
# how의 기본값: inner
pd.merge(df1, df2, on='key', how='inner')
```

```python
df3 = pd.DataFrame({'lkey': list("bbacaab"),
                                   'data1': range(7)})
df4 = pd.DataFrame({'rkey': list("abd"),
                                   'data2': range(3)})
```

```python
df3
```

```python
df4
```

```python
pd.merge(df3, df4, left_on="lkey", right_on="rkey")
```

```python
left1 = pd.DataFrame({'key': ['a', 'b', 'a', 'a', 'b', 'c'], 'value':range(6)})
right1 = pd.DataFrame({'group_val': [3.5, 7]}, index=['a', 'b'])
```

```python
left1
```

```python
right1
```

```python
pd.merge(left1, right1, left_on='key', right_index=True)
```

```python
pd.merge(right1, left1, left_index = True, right_on='key')
```

```python
left2 = pd.DataFrame([[1., 2.], [3. , 4.], [5., 6.]],
                                     index = ['a', 'c', 'e'], columns = ['Seoul', 'Incheon'])
right2 = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [13., 14]],
                                       index = ['b', 'c', 'd', 'e'], columns  = ['Daegu', 'Ulsan'])
```

```python
left2
```

```python
right2
```

```python
pd.merge(left2, right2, how='outer', left_index=True, right_index=True)
```

```python
df1 = pd.DataFrame(np.arange(6).reshape(3, 2),
                                  index = ['a', 'b', 'c'], columns = ['one', 'two'])
df2 = pd.DataFrame(5 + np.arange(4).reshape(2, 2),
                                  index = ['a', 'c'], columns = ['three', 'four'])
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1, df2], axis=1)
```
