```python
import numpy as np
import pandas as pd
```

```python
data = {"names": ["a", "b", "c", "d", "e"],
        "year": [2014, 2015, 2016, 2017, 2018],
        "points": [1.2, 3.2, 3.3, 4.0, 4.1]}
df = pd.DataFrame(data, columns = ['year', 'names', 'points', 'penalty'],
                        index = ['one', 'two', 'three', 'four', 'five'])
```
